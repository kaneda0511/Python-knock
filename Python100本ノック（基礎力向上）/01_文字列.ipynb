# 問題3 文字列の代入(format/f-strings)
以下の文章の「〇〇」に対して`.format()`とf-stringsを使って代入しましょう。  
「私は〇〇です。〇〇出身です。」

（目安時間：1分）


## 解答

````python
name = "金田"
come_from = "栃木"
"私は{}です。出身は{}です。".format(name,come_from)
````

````python
f"私は{name}です。出身は{come_from}です。"
````

# 問題5 書式化

- 符号付き10進数

# 問題9 空白除去

# 問題13 包含関係

- A: `'xdh'`, B: `'orweit'`
